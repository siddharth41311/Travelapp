```python
base_price = 100
def new_price(x): 
    distance_price = 80 * (x['totalDistance']/100)
    if x['carType'] == 'Hatchback':
        return (base_price * 1 * x['rentalDuration'])  + distance_price
    elif x['carType'] == 'Sedan':
        return (base_price * 1.2 * x['rentalDuration']) + distance_price
    elif x['carType'] == 'SUV':
        return (base_price * 1.4 * x['rentalDuration']) + distance_price
    elif x['carType'] == 'Luxury':
        return (base_price * 1.7 * x['rentalDuration']) + distance_price
```

```python
rent_car['Total_car_price'] = rent_car.apply(new_price , axis = 1)
print(rent_car)
```

```python
sid.polarity_scores(hotel_review.iloc[0]['review_hotel'])
```

```python
hotel_review['scores'] = hotel_review['review_hotel'].apply(lambda review: sid.polarity_scores(review))
```

```python
hotel_review['compound'] = hotel_review['scores'].apply(lambda score_dict : score_dict['compound'])
```

```python
hotel_review['comp_score'] = hotel_review['compound'].apply(lambda c: 'pos' if c>0.5 else 'neg' if c<-0.5 else 'neu')
```
